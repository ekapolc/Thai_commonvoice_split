```python
from pythainlp.tokenize import word_tokenize
import pandas as pd
import phunspell

pspell = phunspell.Phunspell('th_TH')
```

```python
cv_root: str = '' #"../data/cv-corpus-7.0-2021-07-21"

train: pd.DataFrame = pd.read_csv("train.tsv", delimiter="\t") # f"{cv_root}/th/train.tsv"
```

```python
sents: list = [word_tokenize(i.strip()) for i in train["sentence"].to_list()]
```

```python
for i in sents:
    _missing = False
    for index, j in enumerate(i):
        if pspell.lookup(j) == False and j != "?" and j != "!":
            _missing = True
            i[index] = "<missing>" + j + "</missing>"
            break
    if _missing:
        print(i)
```
